```python
import pandas as pd
import json
import xml.etree.ElementTree as ET

# --- A. CSV & Excel (XL) ---
def read_tabular_data():
    # Reading CSV
    csv_data = pd.read_csv('data.csv')
    print("--- CSV Data ---")
    print(csv_data.head()) 

    # Reading Excel
    excel_data = pd.read_excel('data.xlsx')
    print("\n--- Excel Data ---")
    print(excel_data.head())

# --- B. JSON (Likely your "350N" data) ---
def read_json_data():
    with open('data.json', 'r') as f:
        data = json.load(f)
    
    # Display nicely
    print("\n--- JSON Data ---")
    print(json.dumps(data, indent=4))

# --- C. XML ---
def read_xml_data():
    tree = ET.parse('data.xml')
    root = tree.getroot()
    
    print("\n--- XML Data ---")
    for child in root:
        print(f"Tag: {child.tag}, Attribute: {child.attrib}")

# read_tabular_data()
# read_json_data()
# read_xml_data()
```

```python
from sqlalchemy import create_engine
import pandas as pd

def collect_from_sql():
    # Replace with your connection string (e.g., postgresql, mysql, sqlite)
    # format: dialect+driver://username:password@host:port/database
    db_connection_str = 'sqlite:///my_database.db'
    db_connection = create_engine(db_connection_str)

    try:
        # Read directly into a Pandas DataFrame
        df = pd.read_sql("SELECT * FROM employees", db_connection)
        print("\n--- SQL Data ---")
        print(df)
    except Exception as e:
        print(f"Database error: {e}")
```
